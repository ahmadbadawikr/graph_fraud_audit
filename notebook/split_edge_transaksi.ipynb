```python
import os
import pandas as pd

from tqdm.notebook import tqdm
```

```python
def parsing_norek(nomor_rekening):
    nomor_rekening = str(nomor_rekening)

    if nomor_rekening[-3:-2] == "5":
        return "simpanan"
    elif nomor_rekening[-3:-2] == "1":
        return "pinjaman"
    else:
        return "lainnya"
```

```python
parquet_dir = "/Users/ymnzaman/Documents/Project/Graph/data/edge_rek_credit"
partitions = os.listdir(parquet_dir)

for partition in tqdm(partitions):
    path = os.path.join(parquet_dir, partition)
    temp = pd.read_parquet(path)

    temp["flag_rekening"] = temp["dst"].apply(lambda x: parsing_norek(x))

    simp = temp[temp["flag_rekening"] == "simpanan"]
    pinj = temp[temp["flag_rekening"] == "pinjaman"]
    del temp

    print("Simpanan: ", simp.shape)
    print("Pinjaman: ", pinj.shape)


    if not os.path.exists("/Users/ymnzaman/Documents/Project/Graph/data/edge_simp_credit"):
        os.mkdir("/Users/ymnzaman/Documents/Project/Graph/data/edge_simp_credit")

    if not os.path.exists("/Users/ymnzaman/Documents/Project/Graph/data/edge_pinj_credit"):
        os.mkdir("/Users/ymnzaman/Documents/Project/Graph/data/edge_pinj_credit")

    simp.to_parquet(f"/Users/ymnzaman/Documents/Project/Graph/data/edge_simp_credit/{partition}")
    pinj.to_parquet(f"/Users/ymnzaman/Documents/Project/Graph/data/edge_pinj_credit/{partition}")

```
